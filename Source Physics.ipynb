```julia
Mej = 0.1*SolarMass; #mass ejected by HN 
n1 = 1.0;            #number density of the surrounding medium [cm^-3]
rho1 = mu*ProtonMass; #mass density of the sorrounding medium [g cm^-3]
E52_5 = 1.0; 
Ekin = E52_5*10^(52.5); 
```

```julia
vs0 = sqrt(2*Ekin/Mej) #initial shock velosity calculated from kinetic energy and ejected mass [cm/s]
@assert(Vs0 < c);
rST = (3*Mej/(4pi*rho1))^(1/3);  #radius that the Sedov-Taylor phase begins [cm]
tST = rST/vs0;                   #time at which Sedov-Taylor phase begins [s] 
t5ST = tST/t5_to_s;        #tST [10^5 yrs] 
```

Calculate the time that the snow plow phase sets in by equating the dynamic time with the time required to cool the plasma from free-free and line cooling (see Peter's notes) 

```julia
T26 = linspace(0.1,10,100000);
tc5 = 200/n1;
temp = 1/sqrt(T26)+45.*T26.^(-1.5);

tc5 = tc5./temp;
td5 = (E52_5/n1)^(1/3)*T26.^(-5/6); 
this_error, SP_index = findmin(abs(tc5-td5))
@assert(this_error < 1e-3)
t5SP = td5[SP_index]        #time at which the snowplow phase begins [10^5 yrs]
```

```julia
t5 = linspace(t5ST,t5SP);  #get a range of times over the Sedov-Taylor phase [10^5 yrs] 
rs = 2e20(E52_5/n1)^(1/5)*t5.^(2/5);  #radius of the shock as a function of time [cm] 
vs = (2/5)*rs./t5/t5_to_s;            #velocity of the shock as a function of time [cm/s]
B = 1.8e-4*sqrt(eps_B_1)*n1^(3/10).*t5.^(-3/5); #magnetic field strength behind the shock [G]
E_p_max = 8.7e15*E52_5^(3/5)*sqrt(eps_B_1)*n1^(-1/10)*t5.^(-4/5); #maxium proton energy as a function of time [eV]
```
